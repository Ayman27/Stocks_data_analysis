```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(maninfra_data['Date'], maninfra_data['Close'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Price in INR', fontsize='11')
plt.title('Maninfra Stock Prices')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(maninfra_data['Date'], maninfra_data['Volume'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Volumes', fontsize='11')
plt.title('Volume Trends', fontsize='20')
plt.grid()
plt.show()
```

```python
maninfra_data['Market Cap'] = maninfra_data['Open']*maninfra_data['Volume']
```

```python
maninfra_data.head()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(maninfra_data['Date'], maninfra_data['Market Cap'], color='orange')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Market Cap', fontsize='11')
plt.title('Market Cap')
plt.grid()
plt.show()
```

```python
maninfra_data.iloc[maninfra_data['Market Cap'].argmax()]
```

```python
maninfra_data['vol'] = (maninfra_data['Close']/maninfra_data['Close'].shift(1)) - 1
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(maninfra_data['Date'], maninfra_data['vol'], color='purple')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Volatility')
plt.grid()
plt.show()
```

```python
maninfra_data['vol'].hist(bins=100, color='orange');
```

```python
maninfra_data['Cumulative Return'] = (1 + maninfra_data['vol']).cumprod()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(maninfra_data['Date'], maninfra_data['Cumulative Return'], color='green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Cumulative Return', fontsize='11')
plt.title('Cumulative Return')
plt.grid()
plt.show()
```

```python
maninfra_data.iloc[maninfra_data['Cumulative Return'].argmax()]
```
