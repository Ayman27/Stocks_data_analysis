```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(tataelxsi_data['Date'], tataelxsi_data['Close'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Price in INR', fontsize='11')
plt.title('Tataelxsi Stock Prices')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(tataelxsi_data['Date'], tataelxsi_data['Volume'])
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Volumes', fontsize='11')
plt.title('Volume Trends', fontsize='20')
plt.grid()
plt.show()
```

```python
tataelxsi_data['Market Cap'] = tataelxsi_data['Open']*tataelxsi_data['Volume']
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(tataelxsi_data['Date'], tataelxsi_data['Market Cap'], color='orange')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Market Cap', fontsize='11')
plt.title('Market Cap')
plt.grid()
plt.show()
```

```python
tataelxsi_data.iloc[tataelxsi_data['Market Cap'].argmax()]
```

```python
tataelxsi_data['vol'] = (tataelxsi_data['Close']/tataelxsi_data['Close'].shift(1)) - 1
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(tataelxsi_data['Date'], tataelxsi_data['vol'], color='purple')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.title('Volatility')
plt.grid()
plt.show()
```

```python
tataelxsi_data['vol'].hist(bins=100, color='orange');
```

```python
tataelxsi_data['Cumulative Return'] = (1 + tataelxsi_data['vol']).cumprod()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(tataelxsi_data['Date'], tataelxsi_data['Cumulative Return'], color='green')
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
ax.set_xlabel('Date', fontsize='11')
ax.set_ylabel('Cumulative Return', fontsize='11')
plt.title('Cumulative Return')
plt.grid()
plt.show()
```
